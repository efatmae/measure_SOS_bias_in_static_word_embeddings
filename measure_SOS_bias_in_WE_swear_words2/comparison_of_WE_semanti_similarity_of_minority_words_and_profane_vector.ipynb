```python
FastText_wiki_subwords_sim_Scores_ethnicity_words_and_profane_words = pd.read_csv("Results/ethnicity_similarities/FastText_wiki_subwords/_minority_and_majority_words_sim_to_profane_vector.csv")
SSWE_sim_Scores_ethnicity_words_and_profane_words = pd.read_csv("Results/ethnicity_similarities/SSWE/_minority_and_majority_words_sim_to_profane_vector.csv")
Debias_w2v_sim_Scores_ethnicity_words_and_profane_words = pd.read_csv("Results/ethnicity_similarities/Debias_w2v/_minority_and_majority_words_sim_to_profane_vector.csv")
P_DESIP_subwords_sim_Scores_ethnicity_words_and_profane_words = pd.read_csv("Results/ethnicity_similarities/P_DESIP/_minority_and_majority_words_sim_to_profane_vector.csv")
U_DESIP_sim_Scores_ethnicity_words_and_profane_words = pd.read_csv("Results/ethnicity_similarities/U_DESIP/_minority_and_majority_words_sim_to_profane_vector.csv")
```

```python
def get_sim_scores(word_list, sim_df):
    w2v_sim = []
    w2v_non_normalized_sim = []
    for i in word_list:
        if i in sim_df["words"].values:
            w2v_sim.append(sim_df[sim_df["words"] == i]["sim_score"].values[0])
            w2v_non_normalized_sim.append(sim_df[sim_df["words"] == i]["cos_sim_scores"].values[0])
        else:
            w2v_sim.append(0)
            w2v_non_normalized_sim.append(0)
    return w2v_sim, w2v_non_normalized_sim
```

```python
harmless_minority_words_clean = ["lesbian","gay","bisexual","transgender","tran","queer",
                "lgbt","lgbtq","homosexual", "woman", "female", "girl","wife","sister","daughter","mother", "non-binary",
               "african", "african american", "asian", "black", "hispanic", "latin", "mexican", "indian", "middle eastern",
                "arab"]
```

```python
len(harmless_minority_words_clean)
```

```python
all_data_dict = {}
words = harmless_minority_words_clean

w2v_sim, w2v_un_normalized_sim = get_sim_scores(harmless_minority_words_clean, w2v_sim_Scores_ethnicity_words_and_profane_words)

glove_wk_sim_scores, glove_wk_un_normalized_sim_scores = get_sim_scores(harmless_minority_words_clean, glove_wk_sim_Scores_ethnicity_words_and_profane_words)

glove_twitter_sim_Scores, glove_twitter_un_normalized_sim_Scores = get_sim_scores(harmless_minority_words_clean, glove_twitter_sim_Scores_ethnicity_words_and_profane_words)

ud_sim_scores, ud_un_normalized_sim_scores = get_sim_scores(harmless_minority_words_clean, ud_sim_Scores_ethnicity_words_and_profane_words)

chan_sim_scores, chan_un_normalized_sim_scores = get_sim_scores(harmless_minority_words_clean, chan_sim_Scores_ethnicity_words_and_profane_words)
```

```python
glove_cc_sim, glove_cc_un_normalized_sim = get_sim_scores(harmless_minority_words_clean, glove_cc_sim_Scores_ethnicity_words_and_profane_words)

glove_cc_large_sim_scores, glove_cc_large_un_normalized_sim_scores = get_sim_scores(harmless_minority_words_clean, glove_cc_large_sim_Scores_ethnicity_words_and_profane_words)

FastText_CC_sim_Scores, FastText_CC_un_normalized_sim_Scores = get_sim_scores(harmless_minority_words_clean, FastText_CC_sim_Scores_ethnicity_words_and_profane_words)

FastText_CC_subwords_sim_scores, FastText_CC_subwords_un_normalized_sim_scores = get_sim_scores(harmless_minority_words_clean, FastText_CC_subwords_sim_Scores_ethnicity_words_and_profane_words)

FastText_wiki_sim_scores, FastText_wiki_un_normalized_sim_scores = get_sim_scores(harmless_minority_words_clean, FastText_wiki_sim_Scores_ethnicity_words_and_profane_words)
```

```python
FastText_wiki_subwords_sim, FastText_wiki_subwords_un_normalized_sim = get_sim_scores(harmless_minority_words_clean, FastText_wiki_subwords_sim_Scores_ethnicity_words_and_profane_words)

SSWE_sim_scores, SSWE_un_normalized_sim_scores = get_sim_scores(harmless_minority_words_clean, SSWE_sim_Scores_ethnicity_words_and_profane_words)

Debias_w2v_sim_Scores, Debias_w2v_un_normalized_sim_Scores = get_sim_scores(harmless_minority_words_clean, Debias_w2v_sim_Scores_ethnicity_words_and_profane_words)

P_DESIP_sim_scores, P_DESIP_un_normalized_sim_scores = get_sim_scores(harmless_minority_words_clean, P_DESIP_subwords_sim_Scores_ethnicity_words_and_profane_words)

U_DESIP_sim_scores, U_DESIP_un_normalized_sim_scores = get_sim_scores(harmless_minority_words_clean, U_DESIP_sim_Scores_ethnicity_words_and_profane_words)
```
