```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
correlations = pd.read_csv("sos_pmi_correlation.csv")
```

```python
correlations.columns
```

```python
correlations_lgtbq = correlations[correlations.SOS=="LGBTQ"]
```

```python
correlations_straight = correlations[correlations.SOS=="Straight"]
```

```python
correlations_women = correlations[correlations.SOS=="WOMEN"]
```

```python
correlations_men = correlations[correlations.SOS=="MEN"]
```

```python
correlations_minor_ethn = correlations[correlations.SOS=="minor_ethnicities"]
```

```python
correlations_white_ethn = correlations[correlations.SOS=="white_ethnicities"]
```

```python
correlations_all_minority = correlations[correlations.SOS=="all_min_words"]
```

```python
correlations_WMC = correlations[correlations.SOS=="WMC"]
```

```python
correlations_COX = correlations[correlations.SOS=="COX"]
```

```python
figure(figsize=(7,5), dpi=1200)
y_axis = np.arange(4)
x_axis = np.arange(-1,1,0.1)
width = 0.1
ind = 0.5
plt.style.use('default')
plt.barh(y_axis ,[correlations_lgtbq[correlations_lgtbq["WE"]=="W2V"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                     correlations_women[correlations_women["WE"]=="W2V"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                     correlations_minor_ethn[correlations_minor_ethn["WE"]=="W2V"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                  correlations_all_minority[correlations_all_minority["WE"]=="W2V"]["correlation_reddit_100K_spearmanr"].values.tolist()[0]
                    ], color='b', label="w2v", height=width)

plt.barh(y_axis+0.1, [correlations_lgtbq[correlations_lgtbq["WE"]=="Glove_WK"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                     correlations_women[correlations_women["WE"]=="Glove_WK"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                     correlations_minor_ethn[correlations_minor_ethn["WE"]=="Glove_WK"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                  correlations_all_minority[correlations_all_minority["WE"]=="Glove_WK"]["correlation_reddit_100K_spearmanr"].values.tolist()[0]
                     ],
         color='r', label="glove-wk", height=width)


plt.barh(y_axis+0.2, [correlations_lgtbq[correlations_lgtbq["WE"]=="Glove_Twitter"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                     correlations_women[correlations_women["WE"]=="Glove_Twitter"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                     correlations_minor_ethn[correlations_minor_ethn["WE"]=="Glove_Twitter"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                  correlations_all_minority[correlations_all_minority["WE"]=="Glove_Twitter"]["correlation_reddit_100K_spearmanr"].values.tolist()[0]
                   ], color='g', label = "glove-twitter", height=width)

plt.barh(y_axis+0.3, [correlations_lgtbq[correlations_lgtbq["WE"]=="UD"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                     correlations_women[correlations_women["WE"]=="UD"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                     correlations_minor_ethn[correlations_minor_ethn["WE"]=="UD"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                  correlations_all_minority[correlations_all_minority["WE"]=="UD"]["correlation_reddit_100K_spearmanr"].values.tolist()[0]
                     ], color='orange', label ="ud", height=width)

plt.barh(y_axis+0.4,[correlations_lgtbq[correlations_lgtbq["WE"]=="Chan"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                     correlations_women[correlations_women["WE"]=="Chan"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                     correlations_minor_ethn[correlations_minor_ethn["WE"]=="Chan"]["correlation_reddit_100K_spearmanr"].values.tolist()[0],
                  correlations_all_minority[correlations_all_minority["WE"]=="Chan"]["correlation_reddit_100K_spearmanr"].values.tolist()[0]
                    ], color='black', label="chan", height=width)


plt.yticks(y_axis, ["LGTBQ","Women","Other ethn.","Marginalized."], fontsize=24, fontweight='bold', rotation=0)
plt.xlabel("correlation score",fontsize=24, fontweight='bold')
plt.xticks(x_axis,fontsize=18, fontweight="bold", rotation=90)
plt.legend(bbox_to_anchor=(0, 1.6), loc='upper left', ncol=2, fontsize=26)
plt.savefig("spearman_correlation_between_sos_and_mean_pmi_stigmatized_group_horizontal_100k.pdf", bbox_inches='tight')
```
