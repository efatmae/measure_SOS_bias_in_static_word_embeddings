```python
import numpy as np
import pandas as pd
from matplotlib import pyplot
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
unfairness_gender_gap_twitter_sexism = pd.read_csv("performance and fairness of models_gender - Twitter_sexism_Test_set.csv")
unfairness_gender_gap_twitter_hate = pd.read_csv("performance and fairness of models_gender - Twitter_Hate_test_set.csv")
unfairness_gender_gap_hate_eval = pd.read_csv("performance and fairness of models_gender - HateEval_test_set.csv")

unfairness_racial_gap_twitter_racism = pd.read_csv("performance and fairness of models_race - Twitter_racism_test_set.csv")
unfairness_racial_gap_twitter_hate = pd.read_csv("performance and fairness of models_race - Twitter_hate_test_Set.csv")
unfairness_racial_gap_hate_eval = pd.read_csv("performance and fairness of models_race - HateEval_test_set.csv")
```

```python
unfairness_gender_gap = pd.concat([unfairness_gender_gap_hate_eval,unfairness_gender_gap_twitter_sexism,unfairness_gender_gap_twitter_hate])
unfairness_racial_gap = pd.concat([unfairness_racial_gap_hate_eval, unfairness_racial_gap_twitter_racism, unfairness_racial_gap_twitter_hate])
```

```python
unfairness_gender_gap_BiLSTM = unfairness_gender_gap[unfairness_gender_gap["Model"] == "BiLSTM"]
unfairness_gender_gap_MLP = unfairness_gender_gap[unfairness_gender_gap["Model"] == "MLP"]

unfairness_racial_gap_BiLSTM = unfairness_racial_gap[unfairness_racial_gap["Model"] == "BiLSTM"]
unfairness_racial_gap_MLP = unfairness_racial_gap[unfairness_racial_gap["Model"] == "MLP"]
```

```python
x_ticks_gender = ["HateEval", "Twiter_Sexism", "Twitter_Hate"]
x_ticks_race = ["HateEval", "Twiter_Racism", "Twitter_Hate"]
```

```python
unfairness_gender_gap_hate_eval[unfairness_gender_gap_hate_eval["Model"]=="LR"]["diff_women_men_FPR"].values
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

x_axis = np.arange(len(x_ticks_gender))
y_ticks = np.arange(-0.2,0.5,0.1)
ytickLabels = ["$%.1f$" % y for y in y_ticks]
width = 0.1

axs[0,0].bar(x_axis,unfairness_gender_gap_BiLSTM[unfairness_gender_gap_BiLSTM.WE =="w2v"]["diff_women_men_FPR"].values, width=width, label="W2V")
axs[0,0].bar(x_axis+0.1,unfairness_gender_gap_BiLSTM[unfairness_gender_gap_BiLSTM.WE =="glove_wk"]["diff_women_men_FPR"].values, width=width, label="Glove-wk")
axs[0,0].bar(x_axis+0.2,unfairness_gender_gap_BiLSTM[unfairness_gender_gap_BiLSTM.WE =="glove_twitter"]["diff_women_men_FPR"].values, width=width, label="Glove-twitter")
axs[0,0].bar(x_axis+0.3,unfairness_gender_gap_BiLSTM[unfairness_gender_gap_BiLSTM.WE =="UD"]["diff_women_men_FPR"].values, width=width, label="UD")
axs[0,0].bar(x_axis+0.4,unfairness_gender_gap_BiLSTM[unfairness_gender_gap_BiLSTM.WE =="Chan"]["diff_women_men_FPR"].values, width=width, label="Chan")
axs[0,0].axhline(unfairness_gender_gap_hate_eval[unfairness_gender_gap_hate_eval["Model"]=="LR"]["diff_women_men_FPR"].values, color='r', linestyle='..')
axs[0,0].set_title("Fairness gender gap (BiLSTM)", fontweight='bold')
axs[0,0].set_ylabel("FPR_women - FPR_men",fontsize=12, fontweight="bold")
axs[0,0].set_xticks([])
axs[0,0].set_yticks(y_ticks)
axs[0,0].set_yticklabels(ytickLabels,fontsize=12, fontweight="bold")

axs[1,0].bar(x_axis,unfairness_gender_gap_MLP[unfairness_gender_gap_MLP.WE =="w2v"]["diff_women_men_FPR"].values, width=width, label="W2V")
axs[1,0].bar(x_axis+0.1,unfairness_gender_gap_MLP[unfairness_gender_gap_MLP.WE =="glove_wk"]["diff_women_men_FPR"].values, width=width, label="Glove-wk")
axs[1,0].bar(x_axis+0.2,unfairness_gender_gap_MLP[unfairness_gender_gap_MLP.WE =="glove_twitter"]["diff_women_men_FPR"].values, width=width, label="Glove-twitter")
axs[1,0].bar(x_axis+0.3,unfairness_gender_gap_MLP[unfairness_gender_gap_MLP.WE =="UD"]["diff_women_men_FPR"].values, width=width, label="UD")
axs[1,0].bar(x_axis+0.4,unfairness_gender_gap_MLP[unfairness_gender_gap_MLP.WE =="Chan"]["diff_women_men_FPR"].values, width=width, label="Chan")
axs[1,0].set_title("Fairness gender gap (MLP)", fontweight='bold')
axs[1,0].set_ylabel("FPR_women - FPR_men",fontsize=12, fontweight="bold")
axs[1,0].set_xticks(x_axis)
axs[1,0].set_xticklabels(x_ticks_gender,fontsize=12, fontweight="bold")
axs[1,0].set_yticks(y_ticks)
axs[1,0].set_yticklabels(ytickLabels,fontsize=12, fontweight="bold")

axs[0,1].bar(x_axis,unfairness_racial_gap_BiLSTM[unfairness_racial_gap_BiLSTM.WE =="w2v"]["diff_ethnicity_white_FPR"].values, width=width, label="W2V")
axs[0,1].bar(x_axis+0.1,unfairness_racial_gap_BiLSTM[unfairness_racial_gap_BiLSTM.WE =="glove_wk"]["diff_ethnicity_white_FPR"].values, width=width, label="Glove-wk")
axs[0,1].bar(x_axis+0.2,unfairness_racial_gap_BiLSTM[unfairness_racial_gap_BiLSTM.WE =="glove_twitter"]["diff_ethnicity_white_FPR"].values, width=width, label="Glove-twitter")
axs[0,1].bar(x_axis+0.3,unfairness_racial_gap_BiLSTM[unfairness_racial_gap_BiLSTM.WE =="UD"]["diff_ethnicity_white_FPR"].values, width=width, label="UD")
axs[0,1].bar(x_axis+0.4,unfairness_racial_gap_BiLSTM[unfairness_racial_gap_BiLSTM.WE =="Chan"]["diff_ethnicity_white_FPR"].values, width=width, label="Chan")
axs[0,1].set_title("Fairness racial gap (BiLSTM)", fontweight='bold')
axs[0,1].set_ylabel("FPR_eth - FPR_white",fontsize=12, fontweight="bold")
axs[0,1].set_xticks([])
axs[0,1].set_yticks(y_ticks)
#axs[0,1].set_yticklabels(ytickLabels,fontsize=12, fontweight="bold")

axs[1,1].bar(x_axis,unfairness_racial_gap_MLP[unfairness_racial_gap_MLP.WE =="w2v"]["diff_ethnicity_white_FPR"].values, width=width, label="W2V")
axs[1,1].bar(x_axis+0.1,unfairness_racial_gap_MLP[unfairness_racial_gap_MLP.WE =="glove_wk"]["diff_ethnicity_white_FPR"].values, width=width, label="Glove-wk")
axs[1,1].bar(x_axis+0.2,unfairness_racial_gap_MLP[unfairness_racial_gap_MLP.WE =="glove_twitter"]["diff_ethnicity_white_FPR"].values, width=width, label="Glove-twitter")
axs[1,1].bar(x_axis+0.3,unfairness_racial_gap_MLP[unfairness_racial_gap_MLP.WE =="UD"]["diff_ethnicity_white_FPR"].values, width=width, label="UD")
axs[1,1].bar(x_axis+0.4,unfairness_racial_gap_MLP[unfairness_racial_gap_MLP.WE =="Chan"]["diff_ethnicity_white_FPR"].values, width=width, label="Chan")
axs[1,1].set_title("Fairness racial gap (MLP)", fontweight='bold')
axs[1,1].set_ylabel("FPR_eth - FPR_white",fontsize=12, fontweight="bold")
axs[1,1].set_yticks(y_ticks)
#axs[1,1].set_yticklabels(ytickLabels,fontsize=12, fontweight="bold")
axs[1,1].set_xticks(x_axis)
axs[1,1].set_xticklabels(x_ticks_race,fontsize=12, fontweight="bold")
```
